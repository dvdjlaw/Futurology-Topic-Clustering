```python
reg = re.compile(r"http[^\s]*",re.IGNORECASE)
lemma = WordNetLemmatizer()
def tokenize(text):    
    text = reg.sub(repl="", string=text)
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalpha()]
    tokens = [word for (word,pos) in pos_tag(tokens) if pos.startswith('NN')]
    tokens = [lemma.lemmatize(word) for word in tokens]
    tokens = [word for word in tokens if not exclude.term.str.contains(word).any()]
    tokens = replace_ec(tokens)
    return tokens

def replace_ec(tokens):
    token_replaced = tokens
    for idx, word in enumerate(tokens):
        if np.sum(replacement["term"]==word) >0:
            token_replaced[idx] = replacement.loc[replacement["term"]==word,"replacement"].values[0]
    return(token_replaced)

tfidfv = TfidfVectorizer(tokenizer=tokenize,max_df=0.05,min_df=0.008,stop_words='english',sublinear_tf=True,norm="l2")
print(df.body_html.head())
tf_matrix = tfidfv.fit_transform(df.body_html)
```

```python
def save_sparse_csr(filename,array):
    np.savez(filename,data = array.data ,indices=array.indices,
             indptr =array.indptr, shape=array.shape )

def load_sparse_csr(filename):
    loader = np.load(filename)
    return csr_matrix((loader['data'], loader['indices'], loader['indptr']),
                         shape = loader['shape'])

save_sparse_csr('tf_idf_m.npz', tf_matrix)
#tf_matrix = load_sparse_csr('tf_idf_m.npz')

print(tf_matrix.shape)
```

```python
def fit_kmeans(matrix,nmax,verbose=False):
    km_fit = dict()
    for i in range(2,nmax):
        km = KMeans(n_clusters=i, init='k-means++', max_iter=100, n_init=12, random_state=1)
        km.fit(tf_matrix)
        km_fit[i] = [metrics.silhouette_score(matrix, km.labels_, metric='euclidean'),
                     metrics.calinski_harabaz_score(matrix, km.labels_),
                     km]
    if(verbose):
        for clust in km_fit:
            print("{0} clusters: {1}".format(clust,km_fit[clust][1]))
    return(km_fit)

km_fit = fit_kmeans(tf_matrix.todense(),20,verbose=True)
```

```python
silhouette = [km_fit[k][1] for k in km_fit.keys()]
plt.plot(silhouette)
#plt.axis([0, 20, 0, 1])
plt.title("calinski-harabaz score for k-means, where 2<k<20")
plt.xticks(range(0,21))
plt.show()
```

```python
# Prepare tokenized BOW for LDA
from gensim import corpora, models

# list for tokenized documents in loop
texts = []

# loop through document list
for i in df.body_html:
    # add tokens to list
    texts.append(tokenize(i))
```

```python
# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)

# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]

nLDA = 6
# generate LDA model
ldamodel = models.ldamodel.LdaModel(corpus, num_topics=nLDA, id2word = dictionary, passes=20)

# Print top 5 terms for each topic
for i in range(0,nLDA):
    #print([[dictionary.id2token[term[0]],term[1]] for term in ldamodel.get_topic_terms(i,5)])
    print([dictionary.id2token[term[0]] for term in ldamodel.get_topic_terms(i,5)])
```
